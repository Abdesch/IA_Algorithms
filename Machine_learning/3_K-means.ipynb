<h1 style="color:red; text-align:center; text-decoration:underline;">Algorithme K-means : Classification non supervisée</h1>


```python
# AVANT : Création de données non classées
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Créer des données simulées avec 3 groupes
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Afficher les données sans clustering
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Données avant clustering")
plt.show()
```

```python
# APRÈS : Application de l'algorithme KMeans
from sklearn.cluster import KMeans

# Appliquer KMeans pour regrouper en 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Affichage des résultats avec les centroïdes
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("Résultat du clustering KMeans")
plt.show()
```

<h3 style="color:#0056b3; text-decoration:underline;">Résultat et Interprétation</h3>

L’algorithme K-Means a été appliqué à un jeu de données bidimensionnel afin de regrouper les points selon leur similarité spatiale.  
Après initialisation des centroïdes, les itérations successives ont permis de former des clusters bien distincts, chaque point étant affecté au centre le plus proche.

La convergence a été atteinte rapidement, avec des clusters compacts et bien séparés, illustrant l’efficacité de K-Means pour des données structurellement simples.  
Ce test met également en évidence l’importance du choix de k, qui influence directement la qualité du regroupement, ainsi que l’intérêt de méthodes telles que le **score de silhouette** ou la **méthode du coude** pour déterminer ce paramètre.

<h1 style="color:red; text-align:center; text-decoration:underline;">Algorithme K-means : Classification non supervis√©e</h1>


```python
# AVANT : Cr√©ation de donn√©es non class√©es
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Cr√©er des donn√©es simul√©es avec 3 groupes
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Afficher les donn√©es sans clustering
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Donn√©es avant clustering")
plt.show()
```

```python
# APR√àS : Application de l'algorithme KMeans
from sklearn.cluster import KMeans

# Appliquer KMeans pour regrouper en 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Affichage des r√©sultats avec les centro√Ødes
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("R√©sultat du clustering KMeans")
plt.show()
```

<h3 style="color:#0056b3; text-decoration:underline;">R√©sultat et Interpr√©tation</h3>

L‚Äôalgorithme K-Means a √©t√© appliqu√© √† un jeu de donn√©es bidimensionnel afin de regrouper les points selon leur similarit√© spatiale.  
Apr√®s initialisation des centro√Ødes, les it√©rations successives ont permis de former des clusters bien distincts, chaque point √©tant affect√© au centre le plus proche.

La convergence a √©t√© atteinte rapidement, avec des clusters compacts et bien s√©par√©s, illustrant l‚Äôefficacit√© de K-Means pour des donn√©es structurellement simples.  
Ce test met √©galement en √©vidence l‚Äôimportance du choix de k, qui influence directement la qualit√© du regroupement, ainsi que l‚Äôint√©r√™t de m√©thodes telles que le **score de silhouette** ou la **m√©thode du coude** pour d√©terminer ce param√®tre.

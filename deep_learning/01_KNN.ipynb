<h1 style="color:red; text-align:center; text-decoration:underline;">Algorithme des k plus proches voisins (k-NN)</h1>


```python
import matplotlib.pyplot as plt
import numpy as np

# Données de catégorie A (turquoise) et B (orange)
category_A = np.array([[1, 2], [1.5, 1.8], [2, 1.5], [2.5, 2], [2.2, 2.5], [1.8, 2.2]])
category_B = np.array([[4, 4], [4.5, 4.2], [5, 4.5], [5.5, 4.8], [5.2, 5], [4.8, 4.7]])

# Point à prédire (non classé)
new_point = np.array([[3, 3]])

# Affichage AVANT KNN
plt.scatter(category_A[:, 0], category_A[:, 1], color='mediumturquoise', label='Category A')
plt.scatter(category_B[:, 0], category_B[:, 1], color='orange', label='Category B')
plt.scatter(new_point[0, 0], new_point[0, 1], color='blue', label='New data point')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Avant KNN - Point non classé')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Regrouper toutes les données
X_train = np.vstack((category_A, category_B))
y_train = np.array([0]*len(category_A) + [1]*len(category_B))  # 0 = A, 1 = B

# Création du modèle KNN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Prédiction
y_pred = knn.predict(new_point)

# Affichage APRÈS KNN
plt.scatter(category_A[:, 0], category_A[:, 1], color='mediumturquoise', label='Category A')
plt.scatter(category_B[:, 0], category_B[:, 1], color='orange', label='Category B')
plt.scatter(new_point[0, 0], new_point[0, 1], color='mediumturquoise' if y_pred == 0 else 'orange',
            edgecolor='k', s=200, label=f'Classé dans catégorie {"A" if y_pred == 0 else "B"}')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Après KNN - Prédiction du point rouge')
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style="color:#0056b3; text-decoration:underline;">Résultat et Interprétation</h3>

L’algorithme k-NN a été appliqué à un jeu de données de classification binaire.  
Avant l’entraînement, la séparation entre les classes n’était pas explicite visuellement, rendant la prédiction incertaine.  
Après l’application de k-NN, le modèle a réussi à affecter correctement les nouvelles observations en fonction de leurs plus proches voisins dans l’espace des caractéristiques.

Le choix du paramètre **k** a influencé la performance :  
- un k trop petit entraîne une forte sensibilité au bruit,  
- un k trop grand peut lisser excessivement les frontières entre classes.

Ce test montre que k-NN est un algorithme efficace et intuitif pour des tâches de classification, à condition de bien ajuster ses paramètres et de normaliser les données si nécessaire.
